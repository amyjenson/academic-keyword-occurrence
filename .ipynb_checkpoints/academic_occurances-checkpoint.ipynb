```python
import matplotlib.pyplot as plt
import os
import pandas as pd

data_path = '/Users/amyjenson/Documents/GitHub/academic-keyword-occurrence/'

data_fn1 = os.path.join(data_path, 'sea_level_rise_glaciers_1950-1994.csv')
data1 = pd.read_csv(data_fn1)
data_fn2 = os.path.join(data_path, 'sea_level_rise_glaciers_1995-2009.csv')
data2 = pd.read_csv(data_fn2)
data_fn3 = os.path.join(data_path, 'sea_level_rise_glaciers_2010-2020.csv')
data3 = pd.read_csv(data_fn3)
data_fn4 = os.path.join(data_path, 'sea_level_rise_glaciers_2021-2024.csv')
data4 = pd.read_csv(data_fn4)
data_fn5 = os.path.join(data_path, 'sea_level_rise_glaciers_1900-1949.csv')
data5 = pd.read_csv(data_fn5)

SLR_glaciers_data = pd.concat([data5, data1, data2, data3, data4], ignore_index=True)

data_fn1 = os.path.join(data_path, 'glaciers_1950-1975.csv')
data1 = pd.read_csv(data_fn1)
data_fn2 = os.path.join(data_path, 'glaciers_1976-2024.csv')
data2 = pd.read_csv(data_fn2)
data_fn3 = os.path.join(data_path, 'glaciers_1900-1939.csv')
data3 = pd.read_csv(data_fn3)
data_fn4 = os.path.join(data_path, 'glaciers_1940-1949.csv')
data4 = pd.read_csv(data_fn4)
glaciers_data = pd.concat([data3, data4, data1, data2], ignore_index=True)

#print(SLR_glaciers_data)
#print(glaciers_data)

SLR_glaciers_year = SLR_glaciers_data['year']
SLR_glaciers_results = SLR_glaciers_data['results']
glaciers_year = glaciers_data['year']
glaciers_results = glaciers_data['results']

plt.plot(SLR_glaciers_year, SLR_glaciers_results)
plt.plot(glaciers_year, glaciers_results)
plt.ylabel('Occurences in academic papers')
plt.xlabel('Date')
plt.legend(['sea level rise glaciers', 'glaciers'])
plt.savefig('occurrences.png')
plt.show()
```

```python
# percentage of total papers published

plt.plot(SLR_glaciers_year, SLR_glaciers_results/glaciers_results)
plt.title('Percentage of total papers published with the word glaciers')
plt.ylabel('Percentage %')
plt.xlabel('Date')
plt.legend(['sea level rise glaciers', 'glaciers'])
plt.savefig('percentage_occurrences.png')
plt.show()
```
